```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
x, y = make_classification(
    n_samples=100, 
    n_features=2, 
    n_informative=1,
    n_redundant=0,
    n_classes=2, 
    n_clusters_per_class=1, 
    random_state=41,
    hypercube=False,
    class_sep=15
)
```

```python
print("X:", x, x.shape)
print("Y:", y, y.shape)
```

```python
plt.figure(figsize = (10,6))
plt.scatter(x[:,0], x[:,1], c=y, cmap='winter', s=100)
```

```python
def perceptron_loss_function(x, y):
    w1 = w2 = b = 1
    learning_rate = 0.01
    epochs = 1000

    for epoch in range(epochs):
        for i in range(x.shape[0]):
            z = w1*x[i][0] + w2*x[i][1] + b

            # Calculating and checking for y_i*f(x_i)
            if (y[i]*z <= 0):
                w1 += learning_rate*y[i]*x[i][0]
                w2 += learning_rate*y[i]*x[i][1]
                b += learning_rate*y[i]

    return w1, w2, b
```

```python
w1, w2, b = perceptron_loss_function(x, y)
print("w1:", w1)
print("w2:", w2)
print("b:", b)
```

```python
m = -(w1/w2)
c = -(b/w2)
print("slope:", m)
print("intercept:", c)
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m*x_input + c

plt.figure(figsize = (10, 6))
plt.plot(x_input, y_input, color='red', linewidth=3)
plt.scatter(x[:,0], x[:,1], c=y, cmap='winter', s=100)
plt.ylim(-3, 2)
```
