```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
from mlxtend.plotting import plot_decision_regions
from sklearn.preprocessing import StandardScaler
```

### Load dataset

```python
df = pd.read_csv('perceptron_dataset.csv')
print(df)
df.head()
```

### Scatter plot with correct column names

```python
sns.scatterplot(x=df['IQ'], y=df['Resume Score'], hue=df['Placed'])
plt.title("Scatter Plot of IQ vs Resume Score (Placement)")
plt.show()
```

### Extract features and labels

```python
x = df.iloc[:, 0:2] # IQ and Resume Score
y = df.iloc[:, -1]  # Placed
```

### Scale features

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

### Train Perceptron

```python
p = Perceptron()
p.fit(x_scaled, y)
```

### Print weights and bias

```python
print("Weights:", p.coef_)
print("Bias (Intercept):", p.intercept_)
```

### Plot Decision Boundary

```python
plot_decision_regions(x_scaled, y.to_numpy(), clf=p, legend=2)
plt.xlabel("IQ")
plt.ylabel("Resume Score")
plt.title("Decision Boundary by Perceptron")
plt.show()
```

### Calculate Accuracy of Perceptron

```python
accuracy = p.score(x_scaled, y)
print("Accuracy (using score method):", accuracy)
```
