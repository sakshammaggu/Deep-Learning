```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Create a 2D dataset
np.random.seed(42)
x = np.random.randn(100, 2) # 100 points with 2 features (x1, x2)
y = (x[:, 0] + x[:, 1] > 0).astype(int) # Labels: 1 if x1 + x2 > 0, else 0
print("X:" , x, x.shape)
print("Y", y, y.shape)
```

```python
# Define two perceptrons with different weights and biases
w1_1, w2_1, b_1 = 1.0, 1.0, -0.5  # Perceptron 1
w1_2, w2_2, b_2 = -0.5, 1.5, 0.2  # Perceptron 2

print("Equation of z1 (Perceptron 1): {:.1f}x1 + {:.1f}x2 + {:.1f} = 0".format(w1_1, w2_1, b_1))
print("Equation of z2 (Perceptron 2): {:.1f}x1 + {:.1f}x2 + {:.1f} = 0".format(w1_2, w2_2, b_2))
```

```python
# Compute decision boundaries
x1_range = np.linspace(min(x[:, 0]), max(x[:, 0]), 100)
x2_1 = (-w1_1 * x1_range - b_1) / w2_1  
x2_2 = (-w1_2 * x1_range - b_2) / w2_2  
```

```python
# Plot only the dataset
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='bwr', alpha=0.7)  
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Dataset Only')
plt.grid(True)
```

```python
# Plot Perceptron 1 Boundary
plt.figure(figsize=(5, 5))
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='bwr', alpha=0.7)
plt.plot(x1_range, x2_1, lw=2, color='green', label='z1=0')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Perceptron 1 Decision Boundary')
plt.legend()
plt.grid(True)
```

```python
# Plot Perceptron 2 Boundary
plt.figure(figsize=(5, 5))
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='bwr', alpha=0.7)
plt.plot(x1_range, x2_2, lw=2, color='orange', label='z2=0')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Perceptron 2 Decision Boundary')
plt.legend()
plt.grid(True)
```

```python
# Plot both decision boundaries with dataset 
plt.figure(figsize=(8, 6))
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='bwr', alpha=0.7)
plt.plot(x1_range, x2_1, lw=2, color='green', label='z1=0')
plt.plot(x1_range, x2_2, lw=2, color='orange', label='z2=0')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Both Perceptron decision Boundaries')
plt.legend()
plt.grid(True)
```
